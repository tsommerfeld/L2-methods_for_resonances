```python
# this is for rac[3,1]
#
#  A=a**2, B=b**2, and D=d**2
#
A, B, D, k, l0 = symbols('A, B, D, k, l0')
```

```python
f = k**2 + 2*A*k + A**2 + B
g = 1 + D*k
h = A**2 + B + 2*A*k + D*A**2*k + D*B*k
y = l0*f*g/h
y
```

```python
diff(y,l0)
```

```python
dA=diff(y,A)
dA.factor()
```

```python
dB=diff(y,B)
dB.factor()
```

```python
dD=diff(y,D)
dD.factor()
```
