```python
# this is for rac[5,3]
#
# f(kappa)
# lambda0(alpha, beta)(gamma,delta)(epsilon)/(...+omega+rho)
#
k, l0, a, b, g, d, e, o, r = symbols('k l0 a b g d e o r', real=True, positive=True)
```

```python
# Roman's new factorization:
f1 = k**2 + 2*a**2*k + a**4+b**2
f2 = 1 + g**2*k + d**2*k**2
f3 = 1 + e**2*k
A = a**4+b**2
B = 2*a**2 + g**2*A + e**2*A

f = l0 * f1 * f2 * f3 / (A + B*k + o**2*k**2 + r**2*k**3)
f
```

```python
diff(f,l0)
```

```python
da=diff(f,a)
da.factor()
```

```python
db=diff(f,b)
db.factor()
```

```python
dg=diff(f,g)
dg.factor()
```

```python
dd=diff(f,d)
dd.factor()
```

```python
de=diff(f,e)
de.factor()
```

```python
do=diff(f,o)
do.factor()
```

```python
dr=diff(f,r)
dr.factor()
```
